```python
import pandas as pd
import numpy as np
```

```python
attributes = [
	'Class',
    'Alcohol',
 	'Malic acid',
 	'Ash',
	'Alcalinity of ash  ',
 	'Magnesium',
	'Total phenols',
 	'Flavanoids',
 	'Nonflavanoid phenols',
 	'Proanthocyanins',
	'Color intensity',
 	'Hue',
 	'OD280/OD315 of diluted wines',
 	'Proline'
]
```

```python
data_path = './wine_dataset/wine.data'
```

```python
df = pd.read_csv(data_path, header=None, names=attributes)
df[::4]
```

```python
df.index
```

```python
def get_outliers_by_group_all_cols(df, group, n=3, output_txt="outliers.txt"):
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    print(f"Columnas consideradas: {numeric_cols}")
    
    bool_df = pd.DataFrame(index=df.index) # reserva 178 rows x 0 cols.

    with open(output_txt, 'w') as file:
        for col in numeric_cols:
            if col == 'Class': continue

            mean = df.groupby(group)[col].transform('mean')
            std = df.groupby(group)[col].transform('std')
        
            mask = (abs(df[col] - mean) >= n * std)
            bool_df[col] = mask 

            file.write(f"\n--- Columna: {col} ---\n")
            file.write(f"Outliers:\n{mask[::4]}\n")
            file.write(f"Total outliers: {mask.sum()}\n")

        filtered_df = df[bool_df.any(axis=1)]
    
    return filtered_df, bool_df
```

```python
filtered_df, bool_matrix = get_outliers_by_group_all_cols(df, 'Class', 2)
```
