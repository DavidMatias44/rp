```python
import pandas as pd
import numpy as np
```

```python
attributes = [
	'Class',
    'Alcohol',
 	'Malic acid',
 	'Ash',
	'Alcalinity of ash  ',
 	'Magnesium',
	'Total phenols',
 	'Flavanoids',
 	'Nonflavanoid phenols',
 	'Proanthocyanins',
	'Color intensity',
 	'Hue',
 	'OD280/OD315 of diluted wines',
 	'Proline'
]
```

```python
data_path = './wine_dataset/wine.data'
```

```python
df = pd.read_csv(data_path, header=None, names=attributes)
df[::4]
```

```python
df.describe
```

```python
def get_norm_by_class(df, method):
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    numeric_cols = [col for col in numeric_cols if col != 'Class']

    df_grouped = df.groupby('Class')

    for col in numeric_cols:
        if col == 'Class': continue

        if method == 'max-scaling':
            df[col] = df_grouped[col].transform(
                lambda x: x / x.max()
            )
        elif method == "min-max":
            df[col] = df_grouped[col].transform(
                lambda x: (x - x.min()) / (x.max() - x.min() if x.max() != x.min() else 1)
            )
        elif method == "z-score":
            df[col] = df_grouped[col].transform(
                lambda x: (x - x.mean()) / x.std() if x.std() != 0 else 0
            )
        elif method == "std-scaling":
            df[col] = df_grouped[col].transform(
                lambda x: x / x.std() if x.std() != 0 else x
            )

    return df

```

```python
df_copy1 = df.copy()
df_copy1 = get_norm_by_class(df_copy1, 'max-scaling')
df_copy1
```

```python
df_copy2 = df.copy()
df_copy2 = get_norm_by_class(df_copy2, 'min-max')
df_copy2
```

```python
df_copy3 = df.copy()
df_copy3 = get_norm_by_class(df_copy3, 'z-score')
df_copy3
```

```python
df_copy4 = df.copy()
df_copy4 = get_norm_by_class(df_copy4, 'std-scaling')
df_copy4
```
