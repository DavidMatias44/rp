```python
import pandas as pd
# from sklearn import tree 
from sklearn.tree import plot_tree, DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./glass.csv')
df
```

```python
df.describe
```

```python
X = df.drop(['ID', 'TYPE'], axis=1)
X
```

```python
y = df['TYPE']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)
```

```python
clf = DecisionTreeClassifier(random_state=23, max_depth=7)
clf.fit(X_train, y_train)
```

```python
plt.figure(figsize=(40, 25))
plot_tree(
    clf,
    feature_names=X.columns,
    class_names=y.astype(str).unique(),
    filled=True,
    rounded=True,
    fontsize=10
)
```

### Se obtienen las importancias de las caracterÃ­sticas.

```python
importancias = clf.feature_importances_
importancias
```

### Se crea un nuevo DataFrame con las importancias.

```python
importancias_df = pd.DataFrame(
    {'feature': X.columns, 'importance': importancias}
)
importancias_df = importancias_df.sort_values(
    'importance', ascending=False
)
importancias_df
```

### Se seleccionan las mejor k caracterÃ­sticas.

```python
k = 4
mejores = importancias_df['feature'].head(k).values
mejores
```

```python
X_mejorado = X[mejores]
X_mejorado
```

```python
svm = SVC(kernel='linear')
```

### Se dividen los datos MEJORADOS. Se entrena y se evalÃºa.

```python
X_train_mejorado, X_test_mejorado, y_train, y_test = train_test_split(X_mejorado, y, test_size=0.2, random_state=23)
```

```python
svm.fit(X_train_mejorado, y_train)
y_pred = svm.predict(X_test_mejorado)
print(f"Error: {1 - accuracy_score(y_pred, y_test)}")
```

### Se dividen los datos NO mejorados. Se entrena y se evalÃºa.

```python
X_train, X_mejorado, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=23)
```

```python
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
print(f"Error: {1 - accuracy_score(y_pred, y_test)}")
```
