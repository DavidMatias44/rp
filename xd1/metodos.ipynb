```python
import pandas as pd
```

```python
attributes = [
	'Class',
    'Alcohol',
 	'Malic acid',
 	'Ash',
	'Alcalinity of ash  ',
 	'Magnesium',
	'Total phenols',
 	'Flavanoids',
 	'Nonflavanoid phenols',
 	'Proanthocyanins',
	'Color intensity',
 	'Hue',
 	'OD280/OD315 of diluted wines',
 	'Proline'
]
```

```python
data_path = './wine_dataset/wine.data'
```

```python
df = pd.read_csv(data_path, header=None, names=attributes)
df
```

## groupby

```python
analisis_por_clase = df.groupby('Class')['Alcohol'].mean()
analisis_por_clase
```

## apply

```python
def get_color_intensity_level(value):
    if value > 7: return 'High'
    else: return 'Low'

color_int_lvl = df['Color intensity'].apply(get_color_intensity_level)
color_int_lvl[::15]
```

## map

```python
def get_alcohol_level(value):
    if value > 13.5:
        return 'High'
    elif 12.5 <= value <= 13.5:
        return 'Medium'
    else:
        return 'Low'

df['Alcohol level'] = df['Alcohol'].map(get_alcohol_level)
df[::20]
```

## transform

```python
def is_high_lvl(value):
    return value > value.quantile(0.75)
    
df['High fenol'] = df.groupby('Class')['Total phenols'].transform(
    is_high_lvl
)
df
```

## aggregate

```python
def mg_to_g(value):
    return value.sum() / 1000

analisis_por_clase = df.groupby('Class').agg({
    'Alcohol': ['mean', 'max'],
    'Proline': mg_to_g 
})
analisis_por_clase
```

## lambda

```python
analisis_por_clase = df.groupby('Class').agg({
    'Alcohol': ['mean', 'max'],
    'Proline': lambda x: x.sum() / 1000 
})
analisis_por_clase
```
