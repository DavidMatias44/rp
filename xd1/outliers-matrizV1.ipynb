```python
import pandas as pd
import numpy as np
```

```python
attributes = [
	'Class',
    'Alcohol',
 	'Malic acid',
 	'Ash',
	'Alcalinity of ash  ',
 	'Magnesium',
	'Total phenols',
 	'Flavanoids',
 	'Nonflavanoid phenols',
 	'Proanthocyanins',
	'Color intensity',
 	'Hue',
 	'OD280/OD315 of diluted wines',
 	'Proline'
]
```

```python
data_path = './wine_dataset/wine.data'
```

```python
df = pd.read_csv(data_path, header=None, names=attributes)
df[::4]
```

```python
test = df.drop('Class', axis=1)
test
```

```python
test.mean()
```

```python
test - test.mean()
```

```python
abs(test - test.mean())
```

```python
n = 2
bool_mat = abs(test - test.mean()) >= n * test.std()
bool_mat[::4]
```

```python
bool_mat[bool_mat['Alcohol'] == True]
```

```python
bool_mat[bool_mat['Flavanoids'] == True]
```

```python
bool_mat[bool_mat['Hue'] == True]
```

```python
cols = df.columns
with open('outliersV2.txt') as file:
    for col in cols:
        if col == 'Class': continue
        file.write(f"{bool_mat[bool_mat[col] == True]}")
```
