```python
import pandas as pd
```

```python
df = pd.read_csv("iris.data", header=None, sep=' ')
df
```

```python
y = df[4] # extraemos la columna de clase
print('Dimensión del vector de clase:',y.shape)
print(y)
```

```python
X= df.drop(4, axis=1) # eliminamos la columna de clase y nos quedamos con el resto
print('Dimensiones de la matriz de datos:',X.shape)
print(X)
```

Impresión de la media del DF para observar que devuelve vector de medias. Además, las operaciones de substracción de DF con escalar o valor absoluto, los resultados son un DF.

```python
print(X.mean())
```

```python
print(X - X.mean())
```

```python
print(abs(X - X.mean()))
```

```python
print(X.std())
```

```python
print(2*X.std())
```

El resultado de la fórmula para detectar 'outliers' es una matriz o DF de booleanos por que
se usa un operador relacional (>=). Esto es los valores que satisfacen la relación son Verdaderos
De lo contrario Falso

```python
n=2
print(abs(X - X.mean()) >= n*X.std())
```

Guardamos la matriz de booleanos para detectar o filtrar quiénes son Verdaderos y por lo tanto
son outliers. Después de encontrarlos se procede a eliminarlos

```python
res = abs(X - X.mean()) >= n*X.std()
print(res)
print(res.sum())
```

La impresión de res.sum() nos indica que la primera y segunda columnas contienen 6 y 5 outliers
respectivamente. Ahora mostramos quiénes son

```python
res[res[0]==True]
```

```python
res[res[1]==True]
```

Ya que sabemos quiénes son los outliers, juntamos el total de muestras booleanas en un nuevo DF
para recuperar sus índices de fila con el fin de eliminarlos.

```python
indOut= res[(res[0]==True) | (res[1]==True)]
indOut
```

Ahora usamos 'index' para devolver los índices del DF booleano y con 'drop' eliminamos dichos índices de 
filas (los outliers) del DF original X

```python
X= X.drop(indOut.index)
X
```

Para verificar que se realizón con éxito la eliminación de outliers, observe que el nuevo DF X
contiene solamente 139 de las 150 muestras originales, por lo que se eliminaron 11 outliers. Además
los primeros dos outliers estaban en las filas 14 y 15, por lo que al mostrar las primeras 20
muestras se puede ver que ya no se encuentran en el DF X modificado (de la fila 13 se salta a la 16).
